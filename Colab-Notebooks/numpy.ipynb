###NumPy Sum

```python
n1 = np.array([10, 20])
n2 = np.array([30, 40])

np.sum([n1, n2])
```

```python
np.sum([n1, n2], axis=0)
```

```python
np.sum([n1, n2], axis=1)
```

```python
#NumPy Comparision
n1 = np.array([10, 20, 30])
n2 = np.array([10, 30, 20])

np.equal(n1, n2)
```

```python
np.array_equal(n1, n2)
```

```python
np.array_equiv(n1, n2)
```

###Joining arrays

```python
n1 = np.array([10, 20, 30])
n2 = np.array([40, 50 ,60])
```

```python
np.vstack((n1, n2))
```

```python
np.hstack((n1, n2))
```

```python
np.column_stack((n1, n2))
```
